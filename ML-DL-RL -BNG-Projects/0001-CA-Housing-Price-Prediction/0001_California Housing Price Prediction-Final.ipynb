#Extract input (X) and output (Y) data from the dataset.

```python
# To get the independent variables, X, drop the dependent variable "median_house_value" from the data frame ca
X=ca.drop('median_house_value',axis=1)
```

```python
X.head()
#but "ocean_proximity" is categorical and needs to be converted to numeric for fitting the machine learning model
```

```python
#Extract the dependent variable y, which is the "median_house_value"

y = ca["median_house_value"]
y.head()
```

# 2. Handle missing values :

```python
X.isnull().sum()
```

```python
#Fill the missing values in the feature "total_bedrooms" with the mean of the respective column.
```

```python
# impute method 

from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
#Fill the missing values for "total_bedrooms" with the mean of the respective column
X['total_bedrooms']=imp.fit_transform(X[['total_bedrooms']])
```

```python
X.isnull().sum() # There are no missing values now for any of the features
```

# 3. Encode categorical data :

```python
#Convert categorical column in the dataset to numerical data.

# Lets first apply value_counts() function on the categorical feature 'ocean proximity"
#Pandas value_counts() function applied to a series returns object containing counts of unique values . 
#The resulting object will be in descending order so that the first element is the most frequently-occurring element.
# type(ca['ocean_proximity']) # is a Pandas series
X['ocean_proximity'].value_counts()
```

```python
#Pandas get_dummies() method converts categorical data into dummy or indicator variables
dummy=pd.get_dummies(X['ocean_proximity'])
dummy.head()
```

#Before we split the data, first concatenate the dummy with X, and remove "ocean_proximity" from the resulting data frame

```python
#concatenate the dummy with X
X=pd.concat([X,dummy],axis=1)
X.head()
```

```python
#Remove "ocean_proximity" from the data frame, so we now have the needed dataframe with dependent variables 

X.drop('ocean_proximity',axis=1,inplace=True)

#When inplace = True , the data is modified in place, which means it will return nothing and the dataframe is now updated. 
#When inplace = False , which is the default, then the operation returns a copy of the object, which then needs to be saved to something.
```

```python
X.head()
```

```python
#Create a new data frame df concatemating the updated independent 'X' variable to the dependent 'y' variable 
df=pd.concat([X,y],axis=1)
```

# 4. Split the dataset :

```python
#Split the data into 80% training dataset and 20% test dataset.

from sklearn.model_selection import train_test_split #Split arrays or matrices into random train and test subsets

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

# 5. Standardize data :

```python
#Standardize training and test datasets.

from sklearn.preprocessing import StandardScaler

#StandardScaler removes the mean and scales each feature/variable to unit variance. 
#It is performed feature-wise in an independent way.
#StandardScaler can be influenced by outliers (if they exist in the dataset) since it 
#involves the estimation of the empirical mean and standard deviation of each feature
```

```python
# scale the independent variables on the training data

scaler=StandardScaler()
```

```python
x_train_sc=scaler.fit_transform(x_train)
```

# 6. Perform Linear Regression : 

```python
#Import LinearRegression method from the sklearn module. 

#linear_model is a class of the sklearn module which contain different functions for performing machine learning with linear models. 
#The term linear model implies that the model is specified as a linear combination of features.

from sklearn.linear_model import LinearRegression
```

```python
# Perform Linear Regression on training data (learn the coefficients)

lm=LinearRegression(fit_intercept=True)

lin_model=lm.fit(x_train_sc,y_train)
```
