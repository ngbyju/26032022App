Accordingly, they do not add any value to the prediction algorithm since the target variable is not affected by the input value,

### Scale the Data

```python
#PCA is effected by scale so we will scale the features in the data before subjecting to PCA. 
#We will transform the data onto unit scale (mean = 0 and variance = 1) which is a requirement for 
#the optimal performance of many machine learning algorithms. StandardScaler helps standardize the dataset’s features. 
# Note that we fit the standardScaler on the training set and transform on the training and test set. 

#scaler = StandardScaler()
## Fit on training set only.
#scaler.fit(Xtrain_mbgf)

## Apply transform to both the training set and the test set.
#Xtrain_mbgf = scaler.transform(Xtrain_mbgf)
#Xtest_mbgf = scaler.transform(Xtest_mbgf)
```

```python
# Following, https://stackoverflow.com/questions/38420847/apply-standardscaler-to-parts-of-a-data-set
# i.e., apply standard scalar only on the eight categorical variables of interest

#First create a copy of your dataframe:

scaled_x = x.copy()
#Don't include the Name column in the transformation:

col_names = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
features = scaled_x[col_names]
#scaler = StandardScaler().fit(features.values)
scaler = MinMaxScaler().fit(features.values)
features = scaler.transform(features.values)
#Now, don't create a new dataframe but assign the result to those two columns:

scaled_x[col_names] = features
#print(scaled_x)
```

```python
#Function to scale label encoded categrical variables of x 
def scale_lex(scaled_x):
    col_names = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
    features  = scaled_x[col_names]
    #scaler   = StandardScaler().fit(features.values)
    scaler    = MinMaxScaler().fit(features.values)
    features  = scaler.transform(features.values)
    scaled_x[col_names] = features
    return scaled_x

scaled_x = x.copy()
scaled_x = scale_lex(scaled_x)
```

### Dimensionality Reduction Employing PCA

```python
scaled_x.shape
```

```python
# Make an instance of the PCA Model

pca=PCA(n_components=7)
pca.fit(scaled_x.iloc[:,0:8])
scaled_x_pca_pc=pd.DataFrame(pca.transform(scaled_x.iloc[:,0:8]),columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
    
# Notice the code below has .95 for the number of components parameter. 
# It means that scikit-learn can choose the minimum number of principal components such that 95% of the variance is retained.
#pca=PCA(0.95) 
```

```python
#Fit PCA on the entire features of training set. Note: We are fitting PCA on the training set only.
#pca.fit(scaled_x)


# Fit PCA only on the eight categorical features of interest (the features we label encoded and scaled in the previous steps):
def fit_pca(df,scaled_x):
    pca=PCA(n_components=7)
    pca.fit(scaled_x.iloc[:,0:8])
    scaled_x_pca_pc=pd.DataFrame(pca.transform(scaled_x.iloc[:,0:8]),columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
    scaled_x_pca_pc=pd.concat([scaled_x_pca_pc,df.ID],axis=1)
    scaled_x_non_pca=pd.concat([scaled_x.iloc[:,8:],df.ID],axis=1)
    scaled_x_pca=pd.merge(left=scaled_x_pca_pc,right=scaled_x_non_pca,on='ID',how='inner')
    scaled_x_pca.drop(columns=['ID'],inplace=True)
    return scaled_x_pca
```

```python
scaled_x_pca=fit_pca(train_mbgf_eor,scaled_x)  # for first parameter, use train_mbgf , if extreme value of y is included; else use train_mbgf_eor 
```

```python
# Note: You can find out how many components PCA choose after fitting the model using pca.n_components_ . 
# In this case, we applied PCA only on 8 features and 99% of the variance, explained among those amounts to 7 principal components 
# After dimensionality reduction, there usually isn’t a particular meaning assigned to each principal component. 


pca.n_components
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.figure(figsize = (15,15))
plt.plot(PC_values, pca.explained_variance_ratio_, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```
