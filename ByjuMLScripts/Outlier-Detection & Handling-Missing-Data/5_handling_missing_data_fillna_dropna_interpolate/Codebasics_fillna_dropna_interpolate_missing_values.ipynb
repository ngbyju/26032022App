```python
import pandas as pd
df=pd.read_csv('weather_data.csv')
df.head()
```

```python
type(df.day[0]) # day is a string
```

```python
#Lets convert it to timestamp data type
df=pd.read_csv('weather_data.csv', parse_dates=["day"])
df.head()
```

```python
type(df.day[0]) #now it is timestamp. Also look at the printed values for day...has changed format
```

```python
df.set_index("day", inplace = True)
```

```python
df # day is not set as index
```

```python
#fill nan with zero
new_df = df.fillna(0)
new_df
```

```python
#But it makes no sense to fill zero for nan's in event feature column
new_df  = df.fillna({'temperature': 0, 'windspeed': 0, 'event': 'no event'})
new_df
```

```python
# It makes no sense to fill nan's in any feature with zero; may be it is more sensible to fillforward
# i.e., fill the values based on what was it yesterday
new_df = df.fillna(method ='ffill')
new_df
```

```python
# It makes no sense to fill nan's in any feature with zero;often it is sensible to fill backward
# i.e., fill the values based on what is it next day
new_df = df.fillna(method ='bfill')
new_df
```

```python
# It makes no sense to fill nan's in any feature with zero; may be it is more sensible to fillforward
# i.e., fill the values based on what was it yesterday
new_df = df.fillna(method ='ffill', limit = 1)  # but carry forward only once... see the difference.. will leave some nan's unchanged
new_df
```

```python
#Interpolate values
# linear interpolation ; works on numerical values and provides better estimates than above methods
new_df = df.interpolate() # default method is linear
new_df
```

```python
#Please note that only method='linear' is supported for DataFrame/Series with a MultiIndex.

#Parameters
#methodstr, default ‘linear’
#Interpolation technique to use. One of:

#‘linear’: Ignore the index and treat the values as equally spaced. This is the only method supported on MultiIndexes.

#‘time’: Works on daily and higher resolution data to interpolate given length of interval.

#‘index’, ‘values’: use the actual numerical values of the index.

#‘pad’: Fill in NaNs using existing values.

#‘nearest’, ‘zero’, ‘slinear’, ‘quadratic’, ‘cubic’, ‘spline’, ‘barycentric’, ‘polynomial’: Passed to scipy.interpolate.interp1d. These methods use the numerical values of the index. Both ‘polynomial’ and ‘spline’ require that you also specify an order (int), e.g. df.interpolate(method='polynomial', order=5).

#‘krogh’, ‘piecewise_polynomial’, ‘spline’, ‘pchip’, ‘akima’, ‘cubicspline’: Wrappers around the SciPy interpolation methods of similar names. See Notes.

#‘from_derivatives’: Refers to scipy.interpolate.BPoly.from_derivatives which replaces ‘piecewise_polynomial’ interpolation method in scipy 0.18.
```

```python
#Interpolate values
# linear interpolation ; works on numerical values and provides better estimates than above methods
new_df = df.interpolate(method = 'quadratic' ) 
new_df
```

```python
new_df = df.interpolate(method ='time')  # a nice method for interolation
new_df
```

```python
# drop all rows with nan

new_df = df.dropna()
new_df
```

```python
# drop only those rows with all features have nan; otherwise dod not delete entire row

new_df = df.dropna(how ="all")  # data from 09 th January is deleted
new_df
```

```python
# Wants to keep rows that has at least one valid value # for instance the row with index 2017-01-06
new_df = df.dropna(thresh = 1)  # thresh =1 means keep the row that has at least one non -nan value
new_df
```

```python
new_df = df.dropna(thresh = 2)  # thresh =2 means keep the row that has at least two valid value
new_df
```

```python
# create rows for all dates from 01-01-2017 to 01-11-2017
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df= df.reindex(idx)
df
```
