```python
#In logistic regression also, we can have  penalty{‘l1’, ‘l2’, ‘elasticnet’, ‘none’}, default=’l2’  
#( not ridge and lasso, but have l1 and l2 norms under logistic regression also; kind of penalty)

# i.e., Specify the norm of the penalty:

#        'none': no penalty is added;

#        'l2': add a L2 penalty term and it is the default choice;

#        'l1': add a L1 penalty term;

#        'elasticnet': both L1 and L2 penalty terms are added.


# Cfloat, default=1.0  (i.e. KN says it is inverse of lambda )
# Inverse of regularization strength; must be a positive float. Like in support vector machines, smaller values specify stronger regularization.

```

```python
# Breast cancer dataset
# housing price dataset
from sklearn.datasets import load_breast_cancer
```

```python
#import requird libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_breast_cancer()
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset.head()
```

```python
# Alternatively
X = pd.DataFrame(df['data'], columns = df['feature_names'])  # Data fllowed by column name
```

```python
X.head()
```

```python
#Dependent Feature
y = pd.DataFrame(df['target'], columns = ['Target']) # Data fllowed by column name
y
```

```python
# Lets first check if y is balanced or imbalanced
y['Target'].value_counts() # Its imbalanced
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
params = [{'C': [1,5,10]}, {'max_iter': [100, 150]}]
```

```python
# logistic model
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(C = 100, max_iter = 100)
```

```python
# logistc model with gridsearch cv
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(model1, param_grid = params, scoring = 'f1', cv = 5)
```

```python
model.fit(X_train, y_train)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred = model.predict(X_test)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
